```
foodData.fillna('NULL', inplace=True)
foodData.info()
```

```
series = foodData.groupby('Food Group').count().index
series2 = []
for items in series:
  itemsNew = items.replace(" ", "")
  series2.append(itemsNew)

series2
```

```
series3 = []

for i in range(len(series)):
  a = pd.DataFrame(foodData[foodData['Food Group'] == series[i]])
  a.reset_index(inplace = True)
  a.drop(columns=['index', 'Food Group','ID'], inplace=True)
  series3.append(a)
```

```
dict = {}
for i in range(len(series)):
  dict[series2[i]] = series3[i]
print(dict)
```

```
for i in list(dict.keys()):
  dict[i].to_csv(index=False,path_or_buf = '{}.csv'.format(i))
```
